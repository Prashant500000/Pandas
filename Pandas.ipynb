```python
import pandas as pd
```

```python
data ={"Name" : ["john", "sita", "shyam"],
      "age" : [21, 22, 23],
      "salary" : [30000, 40000, 50000]
      }
df = pd.DataFrame(data)
print(df)
```

```python
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
print(myvar["y"])
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]}
df = pd.DataFrame(data)

print(df)
```

```python
print(df.loc[0])
```

```python
print(df.loc[[0, 1]])
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df) 
```
